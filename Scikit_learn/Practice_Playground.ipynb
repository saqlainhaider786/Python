# Linear Regression with CSV

```python
data=pd.read_csv("salaries.csv")
data.head()
```

```python
y=data[["salary"]].values
x=data[["years_of_experience"]].values
```

```python
plt.scatter(x,y)
plt.plot()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.4)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
```

```python
from sklearn.metrics import r2_score

acc=r2_score(ytest,ypred)

print("Accuracy Of the model is ",acc*100)
```

# Multi-Linear Regression with CSV

```python
dataset=pd.read_csv("data.csv",usecols=["bedrooms","bathrooms","sqft_living","sqft_above","floors","condition","price"])
dataset.head()
```

```python
dataset = dataset.astype({'bedrooms': int, 'bathrooms': int,"floors":int})
```

```python
dataset.head()
```

```python
dataset.corr()
```

```python
from sklearn.model_selection import train_test_split
y=dataset[["price"]]
x=dataset[["bedrooms","bathrooms","sqft_living","floors","condition","sqft_above"]]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.8)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
```

```python
from sklearn.metrics import r2_score
accuracy=r2_score(ytest,ypred)
accuracy*100
```
