```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=load_iris()
data=iris.data
features=iris.feature_names
target=iris.target_names
```

```python
dataset=pd.DataFrame(data=data,columns=features)
dataset.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
dataset.corr()
```

```python
y=dataset["petal width (cm)"]
x=dataset[["sepal length (cm)","petal length (cm)"]]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
reg.fit(xtrain,ytrain)
```

```python
ypred=reg.predict(xtest)
```

```python
df_pred=pd.DataFrame({"Actual Value ":ytest,"Predicted Value ":ypred})
df_pred
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error

mse = mean_squared_error(ytest,ypred) # A lower Mean Square Error (MSE) indicates better accuracy.
mae = mean_absolute_error(ytest,ypred) # neglects sign of mean of error. Only meant with the absolute value
rmse = np.sqrt(mse) # Root Mean Square Error give value in same unit

print("Mean Absolute Error : ",mae)
print("Mean Squared Error : ",mse)
print("Root Mean Squared Error : ",rmse)
```

```python
data=pd.read_csv("exe2_hiring.csv")
data.head()
```

```python
data
```

```python
data
```
