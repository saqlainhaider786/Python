# CSV Reading using Pandas
***

***Getting csv's complete data***
***

```python
csv_data=pd.read_csv("D:\Python\Data Sets\dermatology.csv")

csv_data
```

***
***Getting csv's specific no. of rows***
***

```python
spcData=pd.read_csv("D:\Python\Data Sets\dermatology.csv",nrows=5)# returns row-1 to row5.

spcData
```

***
***Getting csv's specific no. of cols***
***

```python
spcData=pd.read_csv("D:\Python\Data Sets\dermatology.csv",usecols=[2,0])# returns values of col "2" and "0"

spcData
```

***
***Getting csv's with skipped data***
***

```python
spcData=pd.read_csv("D:\Python\Data Sets\dermatology.csv",skiprows=[0])# returns values of col "2" and "0"

spcData
```
