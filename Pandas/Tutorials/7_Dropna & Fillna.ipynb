```python
import pandas as pd
```

```python
myFil=pd.read_csv("file:///D:\\Python\\Pandas\\PersonsDetails.csv")
print(myFil)
```

***
# Dropna() 
**This method totally works for removing NaN data.**
***

***
**Deleting all rows having NaN values**

```python
myFil.dropna()
```

***
**Applying along any specific axis**

```python
print(myFil)
print()
print(myFil.dropna(axis=1)) #axis 0 for Rows and 1 for Col
```

***
**Deleting any row based on a particular cell's NaN**

```python
print(myFil.dropna(subset=["Full_Name"])) #subset=["col_names"]
print()
print(myFil)
```

***
**Creating a net data set after removing NaN**

```python
print(myFil.dropna(subset=["Full_Name"],inplace=True))
print()
print(myFil)
```

```python
myFil2=pd.read_csv("file:///D:\\Python\\Pandas\\PersonsDetails.csv")
print(myFil2)
```

***
# Fillna()
**This method totally works for inserting data NaN cells.**
***

***
**Inserting New Data in all NaN cells**

```python
myFil2.fillna("Data") # fillna("Data to be filled")
```

***
**Filling New Data within specific column(s)**

```python
myFil2.fillna({"Full_Name":"Ahmed"}) # fillna({col_name:value}) - dictionary is passed
```

```python
# Multiple cols

myFil2.fillna({"Full_Name":"Syed Ali","Ph_Num":"92***********"})
```

***
**Filling NaN cell with previous or next cell's Data**

```python
print(myFil2)
myFil2.ffill() # ffill fills previous data into next empty cell
```

```python
print(myFil2)
myFil2.bfill() # ffill fills next data into previous empty cell
```

***
**Using axis**

```python
print(myFil2)
myFil2.ffill(axis=0) #its default behaviour - ffill fills previous col_data into next empty cell
```

```python
print(myFil2)
myFil2.ffill(axis=1) #ffill fills previous row_data into next empty cell
```

```python
print(myFil2)
myFil2.bfill(axis=0) #its default behaviour - bfill fills next col_data data into previous empty cell
```

```python
print(myFil2)
myFil2.bfill(axis=1) #bfill fills next row_data data into previous empty cell
```

***
**Filling data within a limit**

```python
myFil2.fillna("Nwww",limit=1)#replaces maximum 2 NaN values with "Nwww" in each column
# by default limit works with col 
```
