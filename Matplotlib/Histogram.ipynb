**Histogram**

*A histogram is used to visually represent the distribution of a set of continuous or discrete data, allowing us to see the frequency of data points within specific ranges or bins (also known as intervals or buckets).*

*Ideal for visualizing the distribution of continuous data.*

syntx : plt.hist(var)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np=np.random.randint(10,60,size=50)
```

```python
data=[43,14,38,32,33,58,15,12,48,49,51,30,16,23,34,24,54,52,50,24,16,20,12,28,15,30,30,54,48,56,33,24,52,29,40,45,32,47,40,15,35,17,37,50,51,56,21,51,40,45]
```

```python
plt.title("Histogram")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.hist(data,edgecolor="black")
plt.show()
```

*Adding bins to organize the histogram properly*

```python
ranges=[10,20,30,40,50,60]
plt.title("Histogram")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.hist(data,edgecolor="black",bins=ranges)
plt.show()
```

```python
ranges=[10,20,30,40,50,60]
plt.title("Histogram")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.hist(data,edgecolor="black",bins=ranges,align="right")
plt.show()
```

*Controlling bar width*

```python
ranges=[10,20,30,40,50,60]
plt.title("Histogram")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.hist(data,edgecolor="black",bins=ranges,align="right",rwidth=0.9)
plt.show()
```

*Drawing a horizontal line at any position in the histogram*

```python
ranges=[10,20,30,40,50,60]
plt.title("Histogram")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.hist(data,edgecolor="black",bins=ranges,align="right",rwidth=0.9)
plt.axhline(10,color="red",label="seprater")
plt.legend()
plt.show()
```

*Drawing a vertical line at any position in the histogram*

```python
ranges=[10,20,30,40,50,60]
plt.title("Histogram")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.hist(data,edgecolor="black",bins=ranges,align="right",rwidth=0.9)
plt.axvline(40,color="red",label="seprater")
plt.legend()
plt.show()
```
