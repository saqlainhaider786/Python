***Box Plot***

*Used to visualise max,min and the median*

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=list([[1,2,3,4,5,6],[5,6,4,8,9,7],[3,4,6,1,2,7]])

plt.boxplot(data)
plt.show()
```

*Annotating each point with label*

```python
data=list([[1,2,3,4,5,6],[5,6,4,8,9,7],[3,4,6,1,2,7]])

plt.boxplot(data)
plt.annotate("Min val",xy=(1.1,1.1),xytext=(1.4,1),arrowprops=dict(facecolor="black"))
plt.annotate("Median val",xy=(1.2,3.6),xytext=(1.6,3.5),arrowprops=dict(facecolor="black"))
plt.annotate("Max val",xy=(1.1,6.1),xytext=(1.4,6),arrowprops=dict(facecolor="black",shrink=100))
plt.show()
```
