```python
from matplotlib import pyplot as plt
#import matplotlib.pyplot as plt
```

**Bar Plot**

syntx : plt.bar(x_axis,y_axis)

*A bar plot is a powerful visualization tool primarily used for comparing categorical data or displaying discrete values.*

*Bar plots are  best suited for comparing discrete categories.*
*******

```python
sem=["1st Sem","2nd Sem","3rd Sem","4th Sem"]
cgpa=[85,80,75,90]
```

```python
plt.xlabel("CGPA",fontsize=15)
plt.ylabel("Semesters",fontsize=15)
plt.title("Visualization of CGPA of Semesters",fontsize=15)
plt.bar(sem,cgpa)
plt.show()
```

**Controlling Bar's Width**

```python
plt.xlabel("CGPA",fontsize=15)
plt.ylabel("Semesters",fontsize=15)
plt.title("Visualization of CGPA of Semesters",fontsize=15)
plt.bar(sem,cgpa,width=0.5)
plt.show()
```

**Controlling Bar's Color**

```python
plt.xlabel("CGPA",fontsize=15)
plt.ylabel("Semesters",fontsize=15)
plt.title("Visualization of CGPA of Semesters",fontsize=15)
plt.bar(sem,cgpa,color="green")
plt.show()
```

**Unique color for each Bar**

```python
plt.xlabel("CGPA",fontsize=15)
plt.ylabel("Semesters",fontsize=15)
plt.title("Visualization of CGPA of Semesters",fontsize=15)
colors=["green","khaki","red","orange"]
plt.bar(sem,cgpa,color=colors)
plt.show()
```

**Showing meta data about plot**

```python
plt.xlabel("CGPA",fontsize=15)
plt.ylabel("Semesters",fontsize=15)
plt.title("Visualization of CGPA of Semesters",fontsize=15)
colors=["khaki","green","red","orange"]
plt.bar(sem,cgpa,color=colors,label="result")
plt.legend()
plt.show()
```

**Nested Bar Graph**

```python
import numpy as np

month=["Jan","Feb","Mar","Apr"]
male=[15,4,23,7]
female=[5,5,12,8]

width=0.4
p=np.arange(len(month))
p1=p+width


plt.title("Bar Chart of Selected Candidates of Different Months")
plt.xlabel("Month")
plt.ylabel("Success Candidates")

plt.bar(p,male,width,color="green",edgecolor="black",label="Selected Males")
plt.bar(p1,female,width,color="red",edgecolor="black",label="Selected Females")

plt.legend()
plt.show()
```

*To replace the numerical value on xlabel xticks() is used*

syntx : plt.xticks(position,label)

```python
import numpy as np

month=["Jan","Feb","Mar","Apr"]
male=[15,4,23,7]
female=[5,5,12,8]

width=0.45
p=np.arange(len(month))
p1=p+width


plt.title("Bar Chart of Selected Candidates of Different Months")
plt.xlabel("Month")
plt.ylabel("Success Candidates")

plt.bar(p,male,width,color="green",edgecolor="black",label="Male")
plt.bar(p1,female,width,color="red",edgecolor="black",label="Female")
plt.xticks(p+width/2,month,fontsize=12,rotation=15)

plt.legend(loc="upper right")
plt.show()
```
