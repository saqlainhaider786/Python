```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
l1=np.arange(1,10)
l2=l1*2
```

```python
plt.plot(l1,l2)
```

*Controlling The Line's Style*

```python
plt.plot(l1,l2,linestyle="dotted",linewidth=3)
```

*Multiple Lines*

```python
l3=l1*3
plt.plot(l1,l2,linestyle="dotted",linewidth=3)
plt.plot(l1,l3,linestyle="dashed", color="red")
```

*Using Sub-Plots*

```python
plt.subplot(1,2,1) # 1 x 2 and first plot`
plt.plot(l1,l2)
plt.subplot(1,2,2)
plt.plot(l1,l2)
plt.subplot(2,2,1)
plt.plot(l1,l2)
```
