```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([1,2,3,4,5])
y=np.array([3,1,5,2,6])
plt.plot(x,y)
```

*Adding text info at any position*

```python
x=np.array([1,2,3,4,5])
y=np.array([3,1,5,2,6])
plt.text(0.9,3.5,"Label1")
plt.text(1.9,1.5,"Label2")
plt.plot(x,y)
```

*Text using loop*

```python
for i in range(len(x)):
    st="Label"+str(i+1)
    plt.text(x[i]-0.2,y[i]+0.2,st)
plt.plot(x,y)
```
