```python
import matplotlib.pyplot as plt
```

**Scatter Plot**

syntx : plt.scatter(x_axis,y_axis)

A scatter plot is a versatile visualization tool for exploring relationships between two variables.
***

```python
x=[2,1,4,3,6,5,7]
y=[3,1,2,6,4,7,5]

plt.title("Scatter Graph Example")
plt.xlabel("X_Axis value")
plt.ylabel("Y_Axis value")
plt.scatter(x,y,s=110)
plt.grid()
plt.show()
```

*Working with different colors*

```python
x=[2,1,4,3,6,5,7]
y=[3,9,2,6,4,7,5]
colors=["r","g","r","g","r","g","r"]
plt.title("Scatter Graph Example")
plt.xlabel("X_Axis value")
plt.ylabel("Y_Axis value")
plt.scatter(x,y,c=colors)
plt.show()
```

*working with dots size*

```python
x=[2,1,4,3,6,5,7]
y=[3,9,2,6,4,7,5]
colors=["r","g","r","g","r","g","r"]
size=[55,65,55,65,55,65,55]
plt.title("Scatter Graph Example")
plt.xlabel("X_Axis value")
plt.ylabel("Y_Axis value")
plt.scatter(x,y,c=colors,s=size)
plt.show()
```

*Using cmap*

```python
import numpy as np
x=[2,1,4,3,6,5,7]
y=[3,9,2,6,4,7,5]
colors=np.random.randint(1, 101, size=7)
plt.title("Scatter Graph Example")
plt.xlabel("X_Axis value")
plt.ylabel("Y_Axis value")
plt.scatter(x,y,c=colors,cmap="inferno",s=100,edgecolor="black") #google cmap to see its value's list
cb=plt.colorbar()
cb.set_label("Color bar")
plt.show()
```

*Nested Scatter Plot*

```python
import numpy as np
x=[2,1,4,3,6,5,7]
y=[3,9,2,6,4,7,5]
z=[4,6,9,1,3,8,2]
plt.title("Scatter Graph Example")
plt.xlabel("X_Axis value")
plt.ylabel("Y_Axis value")
plt.scatter(x,y,c="red",s=100,edgecolor="black",label="X&Y") #google cmap to see its value's list
plt.scatter(x,z,c="green",s=100,edgecolor="black",label="X&Z") #google cmap to see its value's list
plt.axhline(5,color="black")
plt.legend()
plt.show()
```

```python
import numpy as np
x=[2,1,4,3,6,5,7]
y=[3,9,2,6,4,7,5]
z=[4,6,9,1,3,8,2]
l1=np.arange(1,8)
l2=l1*2
plt.title("Scatter Graph Example")
plt.xlabel("X_Axis value")
plt.ylabel("Y_Axis value")
plt.scatter(x,y,c="red",s=100,edgecolor="black",label="X&Y") #google cmap to see its value's list
plt.scatter(x,z,c="green",s=100,edgecolor="black",label="X&Z") #google cmap to see its value's list
plt.plot(l1,l2,label="Difference",color="black")
plt.legend()
plt.show()
```
