**Bar Plot Practice Questions**

```python
fav_icecream=["vanila","strawberry","chocolate","others"]
freq=[16,5,12,6]
```

```python
plt.title("Favourite ice-cream flavours poll")
plt.xlabel("ice-cream flavours")
plt.xlabel("frequency")
plt.bar(fav_icecream,freq,color="yellow",alpha=0.6,width=0.5,edgecolor="black")
```

```python
year=["1991-92","1992-93","1993-94","1994-95","1995-96"]
passStu=[150,200,300,250,300]
failStu=[100,100,50,100,100]

width=0.45
interval1=np.arange(len(year))
interval2=interval1+width

plt.title("Result in a School")
plt.xlabel("Year")
plt.ylabel("No. Of Students")

plt.bar(interval1,passStu,width,edgecolor="black",color="green",label="Pass")
plt.bar(interval2,failStu,width,edgecolor="black",color="red",label="Fail")
plt.legend(loc="upper left")
plt.xticks(interval1+width/2,year,rotation=14)
plt.show()
```

```python
subj=["Maths","Science","English","Urdu"]
marks=[75,45,56,80]

plt.title("Student's Marks Bar Chart")
plt.xlabel("marks")
plt.ylabel("subjects")
plt.barh(subj,marks)
```

```python
countries=["North \nAmerica","South \nAmerica","Europe","Asia-Pacific","Africa","Middle East"]
sales=np.array([150,80,120,180,50,70]) #in thousands
info=np.array([sales.max(),sales.min(),sales.sum()])
info
```

```python
plt.title("Energy Drink Sales By Region (Last Quarter")
plt.xlabel("Region")
plt.ylabel("Sales (in thousands)")
for i,value in enumerate(sales):
    plt.text(i,value+5,str(value),ha="center")
plt.bar(countries,sales,color="green",width=0.9,edgecolor="black")

print("Highest Sale : ",info[0],"\nLowest Sale : ",info[1],"\nTotal Sales : ",info[2])
```

*Pie Chart Practice Problems*

```python
expenses=[40000,15000,10000,5000,5000]
categ=["Salaries","Rent","Utilities","Marketing","Miscellaneous"]

plt.pie(expenses,labels=categ,autopct="%0.2f%%",labeldistance=1.05,radius=1.5,textprops={"fontsize":14})
plt.pie([1],colors="white",radius=0.7)
plt.legend(loc="center")
plt.show()
```

```python
values=np.array([60000,35000,25000,15000])
products=["Electronics","Clothing","Home & Kitchen","Beauty & Health"]
plt.pie(values,labels=products,autopct="%0.2f%%",radius=1.5,textprops={"fontsize":14})
plt.legend(loc="upper left")
plt.show()
```

```python
activities = ["School","Homework","Sports and\nExercise","Screen\nTime","Sleep"]
hrs=np.array([8,3,2,4,7])
maxTime=hrs.max()
minTime=hrs.min()
plt.bar(activities,hrs,color="green",edgecolor="black")
x=0
for i in hrs:
    if(i==maxTime):
        plt.text(x,i+0.2,"Highest\nValue",ha="center")
    elif(i==minTime):
        plt.text(x,i+0.2,"Minimum\nValue",ha="center")
    x=x+1

plt.show()
```
