***Use Case :*** Ideal for comparing quantities across different categories.

***Examples :*** Comparing revenue from different regions, product sales, or survey results. Vertical bars show changes over time, while horizontal bars compare quantities at a single point in time.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=sns.load_dataset("penguins")
data.head()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=data)
plt.show()
```

***Differenciating data***

```python
sns.barplot(x="island",y="bill_length_mm",data=data,hue="sex")
plt.show()
```

***Changing bar's order***

```python
order=["Chinstrap","Gentoo","Adelie"]
sns.barplot(x="species",y="bill_length_mm",data=data,hue="island",order=order)
plt.show()
```

```python
order=["Chinstrap","Gentoo","Adelie"]
hueorder=["Dream","Biscoe","Torgersen"]
sns.barplot(x="species",y="bill_length_mm",data=data,hue="island",order=order,hue_order=hueorder)
plt.show()
```

***Color with palette***

```python
sns.barplot(x="island",y="bill_length_mm",data=data,hue="sex",palette="rocket")
plt.legend(loc="upper left")
plt.show()
```

***Setting graph's background***

```python
sns.set(style="darkgrid")
sns.barplot(x="island",y="bill_length_mm",data=data,hue="sex",palette="rocket")
plt.legend(loc="upper left")
plt.show()
```
