```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
var1=[1,2,3,4,5,6,7,8]
var2=[8,7,6,5,4,3,2,1]
```

***Line plot***

```python
sns.lineplot(x=var1,y=var2)
plt.show()
```

*Using as Pandas DataFrame*

```python
data=pd.DataFrame({"x-axis":var1,"y-axis":var2})
sns.lineplot(x="x-axis",y="y-axis",data=data)
plt.show()
```

```python
df1=sns.load_dataset("penguins").head(20)
df1
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=df1)
plt.show()
```

**The above graph has no distribution**

***But to distribute data and make line-graph hue is used***

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=df1,hue="sex")
plt.legend(loc="upper center")
plt.show()
```

*Changing line's color*

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=df1,hue="sex",palette="rocket_r")
plt.legend(loc="upper center")
plt.show()
```

*Adding marker*

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=df1,hue="sex",style="sex",markers=["o",">"])
plt.legend(loc="upper center")
plt.show()
```
