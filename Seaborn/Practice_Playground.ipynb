```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

***Line Plot using seaborn & matplotlib**

```python
loadData=sns.load_dataset("taxis")
loadData
```

```python
data=loadData.iloc[:20,0:]
data
```

```python
sns.lineplot(x="distance",y="fare",data=data)
sns.lineplot(x="distance",y="tip",data=data)
plt.title("Relation B/W Distance & Fare & Tip")
plt.grid()
plt.show()
```

***Histogram***

```python
print(data["total"].min())
print(data["total"].max())
```

```python
sns.displot(data["total"],bins=[5,15,25,35,40],color="red")
plt.show()
```

```python
sns.displot(data["total"],bins=[5,15,25,35,40],color="red",kde=True)
plt.show()
```
