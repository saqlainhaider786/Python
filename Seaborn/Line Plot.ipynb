***Use Case :*** Best for tracking changes over time.

***Examples :*** Showing trends in stock prices, temperature changes, or sales growth over months. Itâ€™s particularly helpful to visualize continuous data, like monthly sales, website traffic, or the progression of a project.

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
var1=[1,2,3,4,5,6,7,8]
var2=[8,7,6,5,4,3,2,1]

lines={"x-axis":var1,"y-axis":var2}
```

```python
sns.lineplot(x="x-axis",y="y-axis",data=lines)
plt.show()
```

***Using External Data set***

```python
x=sns.load_dataset("penguins")
```

```python
x.columns
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=x)
plt.show()
```

***Drawing graph respective to any col/data***

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=x,hue="sex")
plt.show()
```

***Changing line style***

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=x,hue="sex",style="sex")
plt.show()
```

***Changing the line's color***

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=x,hue="sex",palette="Accent")
plt.show()
```

***making the data limited***

```python
x=x.iloc[0:20,0:]
x
```
