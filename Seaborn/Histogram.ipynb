***Use Case:*** Used for showing frequency distributions of continuous data.

***Examples:*** Showing the distribution of test scores, heights of individuals, or ages in a population to identify patterns and outliers.

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
y=sns.load_dataset("penguins")
y.head()
```

```python
# sns.displot(x="bill_length_mm",data=y)
sns.displot(y["bill_length_mm"],bins=[30,35,40,45,50,55,60])
plt.show()
```

***Presenting KDE (Kernal Density Estimator)***

```python
sns.displot(y["bill_length_mm"],bins=[30,35,40,45,50,55,60],kde=True)
plt.show()
```

***Bar Color***

```python
sns.displot(y["bill_length_mm"],bins=[30,35,40,45,50,55,60],kde=True,color="green",alpha=0.5)
plt.show()
```
