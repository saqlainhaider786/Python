```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df1=sns.load_dataset("penguins")
df1
```

**Bar Plot**

```python
sns.barplot(x=df1.island,y=df1.bill_length_mm)
#sns.barplot(x="island",y="bill_length_mm",data=df1)
plt.title("Island VS Panguens Bill Length")
plt.show()
```

*Colors can be changed as in matplotlib*

**Diffenciating data with gender**

```python
sns.barplot(x="island",y="bill_length_mm",data=df1,hue="sex")
plt.title("Island VS Panguens Bill Length")
plt.legend(loc="upper left")
plt.show()
```

**Customizing the category order**

```python
sns.barplot(x="island",y="bill_length_mm",data=df1,hue="sex",order=["Biscoe","Dream","Torgersen"])
plt.title("Island VS Panguens Bill Length")
plt.legend(loc="upper left")
plt.show()
```

*Customizing the bar's order*

```python
sns.barplot(x="island",y="bill_length_mm",data=df1,hue="sex",order=["Biscoe","Dream","Torgersen"],hue_order=["Female","Male"])
plt.title("Island VS Panguens Bill Length")
plt.legend(loc="upper left")
plt.show()
```

*Changing categorized bar color*

```python
sns.barplot(x="island",y="bill_length_mm",data=df1,hue="sex",palette="Accent")
plt.title("Island VS Panguens Bill Length")
plt.legend(loc="upper left")
plt.show()
```
