***Use Case:*** Great for showing relationships or correlations between two variables.

***Examples:*** Examining the relationship between study time and exam scores, height and weight correlation, or age and income data.

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data=sns.load_dataset("penguins")
data.head()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=data,hue="sex")
plt.legend(loc="lower right")
plt.show()
```

***Color***

```python
data=data.iloc[:20,0:]
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=data,hue="sex",palette="cubehelix")
plt.legend(loc="lower right")
plt.show()
```
